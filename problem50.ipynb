```python
import math
import time
start = time.time()

def PrimeGenerator(n):
    primes = [2]
    primeCandidate = 3
    
    # Search for primes till we reach maxInput
    while primeCandidate < n:
        # The limit for checking is the square root of the prime candidate
        testLimit = math.ceil(math.sqrt(primeCandidate))
        
        # Check if the prime is divisable by all previous primes
        check = True
        for prime in primes:
            # Make sure that the check stops if we reach the limit for testing
            if prime <= testLimit:
                if primeCandidate%prime == 0:
                    check = False
            else: 
                break
                
        if check:
            primes.append(primeCandidate)
        primeCandidate +=2
    return primes

def PrimeTest(primeCandidate, primeList):
    # The limit for checking is the square root of the prime candidate
    testLimit = math.ceil(math.sqrt(primeCandidate))

    if primeCandidate == 1:
        return False
    elif primeCandidate == 2:
        return True
    else: 
        # Check if the prime is divisable by all previous primes
        check = True
        for prime in primeList:
            # Make sure that the check stops if we reach the limit for testing
            if prime <= testLimit:
                if primeCandidate%prime == 0:
                    check = False
            else: 
                break

        if check:
            return True
        else:
            return False

# Generate first 1 million primes
primes = PrimeGenerator(1000000)

print('It took', time.time()-start, 'seconds.')
```

```python
primeSum = 0
maxPrime = 1000000

# smallPrimes = [prime for prime in primes if sum(smallPrimes) < maxPrime]
smallPrimes = []
for prime in primes:
    if sum(smallPrimes) < maxPrime:
        smallPrimes.append(prime)
    else:
        smallPrimes.pop()
        break

print('Primes with sum below', sum(smallPrimes))
primeLength = len(smallPrimes)

for consecutiveLength in range(2,primeLength):
    primeSum = []
    for i in range((1+primeLength-consecutiveLength)):
        testPrime = sum(primes[i:i+consecutiveLength])
        if PrimeTest(testPrime, primes):
            print('Consecutive Length:', consecutiveLength, 'Prime Sum:', testPrime)
            
    print(primeSum)
```
